마지막에 빈칸 출력됨

```python
# 상품 이름 xpath
//*[@id="goodsList"]/div[1]/div/ul/li[1]/div/a/span[1] # 처음 상품
//*[@id="goodsList"]/div[1]/div/ul/li[76]/div/a/span[1]
//*[@id="goodsList"]/div[1]/div/ul/li[77]/div/a/span[1]
//*[@id="goodsList"]/div[1]/div/ul/li[99]/div/a/span[1] # 마지막 상품

li[] 숫자만 달라지는 걸 알 수 있음

# 가격 xpath
//*[@id="goodsList"]/div[1]/div/ul/li[98]/div/a/span[2]/span # 할인이 없을 때
//*[@id="goodsList"]/div[1]/div/ul/li[99]/div/a/span[2]/span
//*[@id="goodsList"]/div[1]/div/ul/li[4]/div/a/span[2]/span[2] # span[1]은 할인율

할인율이 있을 때와 없을 때 xpath가 달라져서 복잡하므로 위에서 만든 price_list 사용
```

```python
# xpath를 이용한 상품명 수집
name_list = []

for i in range(1,100): # 처음 인덱스가 1, 마지막 인덱스가 99 이므로
    # 상품 이름 수집
    name_xpath = '//*[@id="goodsList"]/div[1]/div/ul/li['+str(i)+']/div/a/span[1]'
    name_list.append(driver.find_element_by_xpath(name_xpath).text) # 상품명 
```

```python
price_list.pop() # 가격 리스트 마지막 빈칸 제거
```

```python
item = pd.DataFrame( {'name' : name_list,
                      'price' : price_list})
item
```

## 4. 첫 번째 페이지부터 세 번째 페이지까지 상품명 수집

```python
total_name=[]
for page in range(1,4):
    page_path = '//*[@id="goodsList"]/div[2]/div/span['+str(page)+']/a'
    driver.find_element_by_xpath(page_path).click()
    for goods in range(1,99):
        name_x_path = '//*[@id="goodsList"]/div[1]/div/ul/li['+str(goods)+']/div/a/span[1]'
        total_name.append(driver.find_element_by_xpath(name_x_path).text)
```

## 6. 이미지 수집

```python
from urllib.request import urlretrieve

img_urls=[]
for i in range(1,100):
    image_path = '//*[@id="goodsList"]/div[1]/div/ul/li['+str(i)+']/div/div/a/img'
    img_url = driver.find_element_by_xpath(image_path).get_attribute('src') # 이미지 url 수집
    urlretrieve(img_url, f"./test_image/{i}.jpg") # test_imasge 폴더에 저장
    img_urls.append(img_url)
```
