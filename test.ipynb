```python
#마켓컬리정보.json 이름 변경
import json
import re
with open('/Users/hyojin/Mealkit_Recommendation/data/마켓컬리정보.json') as f:
    json_data  = json.load(f)
opt = " \(.*\)|\[.*\] |\(.*\)| \d{3}|[g]"

for i in range(len(json_data)):
    json_data[i]['name'] = re.sub(opt,"",json_data[i]['name'])

with open('/Users/hyojin/Mealkit_Recommendation/data/마켓컬리정보.json','w', encoding='utf-8') as make_file:
    json.dump(json_data, make_file, indent="\t",ensure_ascii=False)
```

```python
import pandas as pd
import numpy as np
import json

with open('/Users/hyojin/Mealkit_Recommendation/data/마켓컬리리뷰_1024_drop_duplicates.json') as f:
    data = json.load(f)

    # print(len(data))
    # print(data[1]['reviews'][0]['user'])
    # print(len(data[1]['reviews']))

    names = []
    users = []
    stars = []
    contents = []

    kurly_reviews = pd.DataFrame(columns=['name', 'user','star','content'])

    for i in range(len(data)):
        for j in range(len(data[i]['reviews'])):
            names.append(data[i]['name'])
            users.append(data[i]['reviews'][j]['user'])
            stars.append(data[i]['reviews'][j]['star'])
            contents.append(data[i]['reviews'][j]['content'])
    
    kurly_reviews['name'] = names
    kurly_reviews['user'] = users
    kurly_reviews['star'] = stars
    kurly_reviews['content'] = contents
```
